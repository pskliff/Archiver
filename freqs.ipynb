```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Freqs

```python
freqs = pd.read_csv('cmake-build-debug/freqs.csv')
```

```python
entropy = pd.read_csv('csvs/entropy.csv')
```

```python
plt.figure(figsize=(15,10))
plt.ylim((0, 0.023 ))

sns.boxplot(x='name', y='freq', data=freqs, fliersize=0)
sns.pointplot(x='name', y='entropy', data=cust_entropy, label='Entropy', color='grey')
plt.legend(['( / 90) + 0.01'])
plt.xlabel(" ")
plt.ylabel('   ')
# plt.legend()
plt.savefig('graphs/file_freq_boxplot.png', dpi=200)
plt.show()
```

```python
grp = freqs.groupby('name')
```

```python
# d1 = freqs[freqs['name'] == 1]
# d2 = freqs[freqs['name'] == 2]
# d3 = freqs[freqs['name'] == 3]
# d4 = freqs[freqs['name'] == 4]
n1 = 1
n2 = 2 
n3 = 3
n4 = 4
d1 = grp.get_group(1)
x1 = d1['char']
y1 = d1['freq']
d2 = grp.get_group(2)
x2 = d2['char']
y2 = d2['freq']
d3 = grp.get_group(3)
x3 = d3['char']
y3 = d3['freq']
d4 = grp.get_group(4)
x4 = d4['char']
y4 = d4['freq']
```

```python
fig, ax = plt.subplots(2, 2, sharex=True, figsize=(15,10))
ax[0][0].plot(x1,y1)
ax[0][0].set_title(' 01')
ax[0][0].set_xlabel('Char')
ax[0][0].set_ylabel('Frequency')

ax[0][1].plot(x2,y2)
ax[0][1].set_title(' 02')
ax[0][1].set_xlabel('Char')
ax[0][1].set_ylabel('Frequency')

ax[1][0].plot(x3,y3)
ax[1][0].set_title(' 03')
ax[1][0].set_xlabel('Char')
ax[1][0].set_ylabel('Frequency')

ax[1][1].plot(x4,y4)
ax[1][1].set_title(' 04')
ax[1][1].set_xlabel('Char')
ax[1][1].set_ylabel('Frequency')
plt.show()
fig.savefig('graphs/f' + str(n1) + str(n2) + str(n3) + str(n4) + '.png', dpi=200)
```
